```python
import numpy as np
```

```python
data=np.random.rand(5)*100
```

```python
data
```

```python
def MinMaxNormalizasyonu(data):
    minVeri=np.min(data)
    maxVeri=np.max(data)
    sonuc=(data-minVeri)/(maxVeri-minVeri)
    return sonuc
# verileri 0-1 arasina normalize eder
```

```python
normalizeEdilmisData=MinMaxNormalizasyonu(data)
print("Normalizasyon oncesi:",data)
print("********************************************************************************")
print("Normalizasyon sonrasi:",normalizeEdilmisData)
```

```python
def IstatikselNormalizasyon(data):
    ortalama=np.min(data)
    standartSapma=np.std(data)
    sonuc=(data-ortalama)/(standartSapma)
    return sonuc
```

```python
normalizeEdilmisData2=IstatikselNormalizasyon(data)
print("Normalizasyon oncesi:",data)
print("********************************************************************************")
print("Normalizasyon sonrasi:",normalizeEdilmisData2)
```

```python
def logaritmikNormalizasyon(data):
    return np.log(data)
```

```python
normalizeEdilmisData3=logaritmikNormalizasyon(data)
print("Normalizasyon oncesi:",data)
print("********************************************************************************")
print("Normalizasyon sonrasi:",normalizeEdilmisData3)
```

```python
def karekokNormalizasyonu(data):
    return np.sqrt(data)
```

```python
normalizeEdilmisData4=karekokNormalizasyonu(data)
print("Normalizasyon oncesi:",data)
print("********************************************************************************")
print("Normalizasyon sonrasi:",normalizeEdilmisData4)
```
