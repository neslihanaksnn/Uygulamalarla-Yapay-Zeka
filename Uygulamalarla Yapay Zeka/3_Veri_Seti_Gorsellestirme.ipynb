```python
Versicolor=veri[veri.variety=="Versicolor"]
Virginica=veri[veri.variety=="Virginica"]
Setosa=veri[veri.variety=="Setosa"]
```

```python
plt.plot(Setosa["petal.length"])
plt.xlabel("id")
plt.ylabel("petal.length")
plt.title("setosa sınıfının görselleştirilmesi")
plt.show()
```

```python
plt.plot(Virginica["petal.length"])

plt.xlabel("id")
plt.ylabel("petal.length")
plt.title("Virginica sınıfının görselleştirilmesi")
plt.show()
```

```python
plt.plot(Versicolor["petal.length"])

plt.xlabel("id")
plt.ylabel("petal.length")
plt.title("Versicolor sınıfının görselleştirilmesi")
plt.show()
```

```python
plt.plot(Setosa["petal.length"],color="blue",label="Setosa")
plt.plot(Versicolor["petal.length"],color="orange",label="Versicolor")
plt.plot(Virginica["petal.length"],color="green",label="Virginica")
plt.legend()

plt.xlabel("id")
plt.ylabel("petal.length")
plt.title("Veri sınıflarının görselleştirilmesi")
plt.show()
```
