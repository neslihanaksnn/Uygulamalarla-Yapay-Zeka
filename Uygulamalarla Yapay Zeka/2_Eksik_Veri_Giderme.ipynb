Dublicate Methodu

```python
veri=pd.DataFrame({'kişi':["mete","mete","ferhat","ferhat","batu","batu","batu"],
                  'kesit':["a","a","b","b","c","c","c"],
                  'sonuc':[1,1,0,0,0,2,2]})
```

```python
veri
```

```python
veri.duplicated() #aynı veriden varsa true yoksa false dondurur
```

```python
veri.drop_duplicates() #aynı olan verileri kaldırır
```
